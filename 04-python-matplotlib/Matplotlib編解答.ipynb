### 8. 凡例  
以下のコードを実行したグラフに対して、sin,cosという凡例を付けたグラフを作成してください。

```python
x = np.linspace(0, 2*np.pi, 500)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1)
plt.plot(x, y2)

plt.ylim(-2, 2)
```

### 15. 3次元グラフ  
次のデータを使って3次元に点をプロットしてみてください。  
こちらのデータはz軸方向に螺旋のような動きをしながら伸びていくものです。

```python
theta = np.linspace(-4*np.pi, 4*np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)
```

```python
# -------解答-------

from mpl_toolkits.mplot3d import Axes3D

theta = np.linspace(-4*np.pi, 4*np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)


fig = plt.figure() # 3Dグラフを描画するためにはfigureを作成する必要がある
ax = fig.add_subplot(111, projection='3d') # 3Dグラフを描画するためにはadd_subplotにprojection='3d'を指定する必要がある

ax.plot(x, y, z, label='3D Curve')
ax.legend() # 凡例を表示する

plt.show()
```

## 17.GUIウィジェット  
以下のコードで作成した散布図上にカーソルを持っていくとカーソルの座標が表示されるようにしてください。  
ただし、MatplotlibのGUIウィジェットはJupyter Notebook上では使えないため、  
この問題に関してはターミナルでPythonインタプリタを起動した状態で確認してください。

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, facecolor='#FFFFCC') # グラフの背景色を指定

x, y = 4*(np.random.rand(2, 100)- .5) # 100個の乱数を生成
ax.plot(x, y, 'o') # 散布図を描画
ax.set_xlim(-2, 2) # x軸の範囲を指定
ax.set_ylim(-2, 2) # y軸の範囲を指定

#下記コードに追加して作成
plt.show()
'''
ヒント
今回はMatplotlibのウェジェットを使うためにmatplotlib.widgets
をモジュールの中Cursorクラスを利用します。
'''
```

```python
# -------解答-------

import matplotlib.pyplot as plt
import numpy as np

from matplotlib.widgets import Cursor

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, facecolor='#FFFFCC')

x, y = 4*(np.random.rand(2, 100) - .5)
ax.plot(x, y, 'o')
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)

Cursor = Cursor(ax, useblit=True, color='red', linewidth=2)

plt.show()
```

### 20.演習問題  
次の3つのデータを使って、以下の図のような積み上げタイプのヒストグラムを作成してください。ただし、ビンの数は10としています。

```python
mu1, sigma1 = 100, 15 # 平均と標準偏差
mu2, sigma2 = 90, 20 # 平均と標準偏差
mu3, sigma3 = 110, 10 # 平均と標準偏差
x1 = mu1 + sigma1 * np.random.randn(100) # 正規分布に従う乱数を100個生成 
x2 = mu2 + sigma2 * np.random.randn(100) # 正規分布に従う乱数を100個生成
x3 = mu3 + sigma3 * np.random.randn(100) # 正規分布に従う乱数を100個生成
```

```python
# -------解答-------

mu1, sigma1 = 100, 15
mu2, sigma2 = 90, 20
mu3, sigma3 = 110, 10
x1 = mu1 + sigma1 * np.random.randn(100) # 平均mu1, 標準偏差sigma1の正規分布に従う乱数を100個生成
x2 = mu2 + sigma2 * np.random.randn(100) # 平均mu2, 標準偏差sigma2の正規分布に従う乱数を100個生成
x3 = mu3 + sigma3 * np.random.randn(100) # 平均mu3, 標準偏差sigma3の正規分布に従う乱数を100個生成

plt.hist([x1, x2, x3], bins=10, color=['yellow', 'skyblue', 'blue'], label=['x1', 'x2', 'x3'], stacked=True)
plt.title('Histogram') # グラフのタイトルを指定
plt.xlabel('x')
plt.ylabel('freq')
plt.xlim(0, 180)
plt.grid(True)
plt.legend(loc='upper left')
plt.show()

```
